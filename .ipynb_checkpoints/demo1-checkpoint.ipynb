```python
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
from pylatexenc import *
import matplotlib.pyplot as plt
```

```python
# one qubit and a classical bit each associated, since a qubit collapses to a state upon measurement
# the collapsed state is what is held by the classical bit
qc = QuantumCircuit(1,1) 

# Hadamard(|0>) --> (|0> + |1>)/2**(0.5) # 
qc.h(0) # applies the Hadamard gate
qc.x(0)
qc.h(0) # applies the Hadamard gate


# measure the value in the qubit
qc.measure(0,0)
```

```python
sim = AerSimulator()
tqc = transpile(qc, sim)
display(qc.draw('mpl'))
result = sim.run(tqc, shots = 2000).result() # shots: the number of iterations to measure the occurences, just like in a coin toss exp.
counts = result.get_counts()
print(counts)
# optional: plot a histogram
plot_histogram(counts)
```
