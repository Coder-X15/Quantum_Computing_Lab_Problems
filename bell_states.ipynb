## Problem:
Create the simulation of the four Bell states (EPR pairs) using Qiskit:
1. $$\mid \phi^+ \rangle = \frac{\mid 00 \rangle + \mid 11 \rangle}{\sqrt{2}}$$
2. $$\mid \phi^- \rangle = \frac{\mid 00 \rangle - \mid 11 \rangle}{\sqrt{2}}$$
3. $$\mid \psi^+ \rangle = \frac{\mid 01 \rangle + \mid 10 \rangle}{\sqrt{2}}$$
4. $$\mid \psi^- \rangle = \frac{\mid 01 \rangle - \mid 10 \rangle}{\sqrt{2}}$$

```python
# Setup
!pip -q install qiskit qiskit-aer
!pip -q install pylatexenc

# Imports
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
from pylatexenc import *
import matplotlib.pyplot as plt
```

```python
# simulation function
def simulate(qc: QuantumCircuit) -> None:
    sim = AerSimulator()
    tqc = transpile(qc, sim)
    display(qc.draw('mpl'))
    result = sim.run(tqc, shots = 2000).result() # shots: the number of iterations to measure the occurences, just like in a coin toss exp.
    counts = result.get_counts()
    return counts
```

## Solution to 1:

```python
phi_plus = QuantumCircuit(2, 2)
phi_plus.h(0) # applying a Hadamard at the 0th qubit.
phi_plus.cx(0, 1) # applying a CNOT to add entanglement
phi_plus.measure([0,1], [0,1]) # measuring

counts = simulate(phi_plus)
```

```python
plot_histogram(counts)
```

## Solution to 2

```python
phi_minus = QuantumCircuit(2, 2)
phi_minus.x(0) # applying a Pauli-X gate to add in a relative phase difference of pi
phi_minus.h(0) # applying a Hadamard at the 0th qubit.
phi_minus.cx(0, 1) # applying a CNOT to add entanglement
phi_minus.measure([0,1], [0,1]) # measuring

counts = simulate(phi_minus)
```

```python
plot_histogram(counts)
```

## Solution to 3

```python
psi_plus = QuantumCircuit(2, 2)
psi_plus.h(0) # applying a Hadamard at the 0th qubit.
psi_plus.x(1) # applying a Pauli-X gate to add in a relative phase difference of pi
psi_plus.cx(0, 1) # applying a CNOT to add entanglement
psi_plus.measure([0,1], [0,1]) # measuring

counts = simulate(psi_plus)
```
