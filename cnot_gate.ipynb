```python
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
from pylatexenc import *
import matplotlib.pyplot as plt
```

```python
qc = QuantumCircuit(2,2) 
# since CNOT is a conditional gate with the first qubit indicating the switch and the second indicating the input,
# we need a 2-qubit system
```

## Todo:
1. Apply a Pauli-X gate on each of the qubits 0 and 1 and verify the functionality of the CNOT gate operation

```python
# comment the below Hadamard gates out to see 00 state appreaing 2000 times :)
qc.h(0) # apply the Hadamard at qubit 0
qc.h(1) # apply the Hadamard at qubit 1
qc.cx(0, 1) # apply the CNOT gate (control bit : 0, target bit: 1)
qc.measure([0,0], [1,1])
```

```python
sim = AerSimulator()
tqc = transpile(qc, sim)
display(qc.draw('mpl'))
result = sim.run(tqc, shots = 2000).result() # shots: the number of iterations to measure the occurences, just like in a coin toss exp.
counts = result.get_counts()
print(counts)
# optional: plot a histogram
plot_histogram(counts)
```
